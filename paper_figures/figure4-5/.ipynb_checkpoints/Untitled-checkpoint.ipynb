```python
import glob

import pandas as pd
```

```python
# files = glob.glob("../../../results/tcga/RECON3D/tcga_Breastv2/*")
files = glob.glob("../../../results/tcga/RECON3D/tcga_Kidneyv2/")
```

```python
df = pd.DataFrame()
for i, file in enumerate(files):
    if "summaryTable" in file:continue
    csv = pd.read_csv(file, sep="\t", lineterminator="\n")
    sample = file.split("/")[-1].split(".")[0]
    fluxes = csv["0"]
    
    if i == 0:
        df["Reaction"] = csv["Unnamed: 0"]
        df[sample] = fluxes
    else:
        df[sample] = fluxes
        
df = df.set_index("Reaction")
df.to_csv("Kidney_fluxesv2.csv")
```
